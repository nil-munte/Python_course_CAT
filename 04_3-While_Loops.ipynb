# While Loops

A diferència dels bucles **for**, els bucles **while** són *indefinits*, és a dir, que s'utilitzen quan no se sap amb exactitud quantes repeticions s'han de realitzar. Per tant, finalitzen quan es compleix una certa **condició**.

L'estructura del bucle **while** és la següent:

**while** *condició:*<br> 
&nbsp;&nbsp;&nbsp;&nbsp;*codi a executar*<br>

Característiques:
- El bucle while comença amb la paraula **while**.
- Seguidament, hi trobem una **condició**, de la mateixa manera que vam veure amb els condicionals: una variable booleana (True o False), una igualtat entre variables (a==b), una condició de desigualtat entre variables o altres números (a < b, a >= 9), etc.
- Al final de la línia, hi apareixen els **dos punts (:)**.
- En el codi a executar, interior al bucle, s'ha de **resoldre la condició**, és a dir, modificar el valor de les variables de la condició perquè pugui finalitzar el bucle. En cas contrari, resultarà en un bucle infinit!
- El codi de dins el bucle ha d'estar **tabulat**, tal com féiem en el bucle for.

## Exemple

Una aplicació molt habitual dels bucles while és fer el que es coneix com una **cerca**. Donada una llista, volem saber quina posició ocupa una paraula.

El següent bucle while depèn de dues condicions: 
- El comptador i, que compta fins el total d'elements de la llista. S'inicialitza a 0.
- El booleà 'trobat', que s'inicialitza a *False* i es posa a *True* si es troba l'element a la llista.

L'existència d'aquestes dues paraules permet sortir del bucle quan es troba la paraula o, en cas que la paraula no sigui a la llista, sortir del bucle quan s'ha recorregut tota la llista.

La variable 'trobat' és *False* a l'inici. Al posar un *not* davant, aquesta condició serà certa. La variable i és sempre més petita que la longitud de la llista (fins que no l'ha recorregut tota). Per tant, a l'inici seguirà sempre amb el bucle. Quan 'trobat' es converteixi a *True*, o es recorri tota la llista (i = len(fruites)), una de les dos condicions es tornarà a *False* i es sortirà del bucle. No es fa servir un *or* perquè requeriria que les dues condicions fossin *False* i no és necessari.

```python
fruites = ["poma", "pera", "plàtan", "préssec", "mandarina", "pruna", "maduixa", "albercoc", "síndria", "meló"]

trobat = False
i = 0

while (not trobat) and (i < len(fruites)):
    if (fruites[i] == "maduixa"):
        trobat = True
    else:
        i+=1

print("La maduixa està a la posició",i,"de la llista de fruites.")
```

# For Loops vs. While Loops

S'utilitzen els **bucles for** quan:
- El nombre d'interacions és conegut o finit (utilitzant **range()**).
- Quan es vol iterar sobre una **llista** o **dictionary**.

S'utilitzen els **bucles while** quan:
- Les iteracions han de seguir fins que es compleixi una **condició**.

Tot i que moltes vegades la funcionalitat és la mateixa utilitzant un **for** que un **while**, es recomana seguir els consells anteriors.
