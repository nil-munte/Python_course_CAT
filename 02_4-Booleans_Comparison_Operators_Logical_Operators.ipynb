# Booleans, Comparison Operators and Logical Operators

## Comparison operators

Sovint és necessari comparar diferents números i variables entre ells per tal de treballar amb aquest resultat en futures operacions. Aquests operadors són:
- (<) Més petit que
- (>) Més gran que
- (==) Igual que
- (!=) Diferent que
- (<=) Més petit o igual que
- (>=) Més gran o igual que

Serveixen per comparar números (o variables que hagin près valor d'un número). El resultat de les comparacions serà un booleà, és a dir, **True** o **False**.

**ALERTA!** L'operador '=' serveix per assignar valor a les variables mentre que, l'operador '==' serveix per comparar si dues variables són iguals.

### Exemple

```python
x = 42 > 43
print(x)

age = 14
is_teen = age > 12
print(is_teen) # True
```

## Logical operators

Existeixen també tres operadors lògics. Serveixen per realitzar comparacions quan es voler satisfer dues o més condicions a la vegada.

Abans, serà necessari repassar les taules de la veritat AND i OR: https://docs.oracle.com/html/E79061_01/Content/Reference/Truth_tables.htm

- (and) Evalua si totes les condicions són **True**. En cas que una sigui **False**, el resultat total serà **False**.
- (or) Evalua si almenys **una** condició és **True**. En cas que només una sigui **True**, el resultat total serà **True**.

### Exemple

```python
age = 14
is_teen = age > 12 and age < 20
print(is_teen) # True
```
