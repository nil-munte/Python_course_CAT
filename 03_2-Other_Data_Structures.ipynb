# Other Data Structures

## List of lists

Una **llista de llistes** ('list of lists') és una llista on, cada element, és també una llista. És una estructura bidimensional (2D list). Tindria la forma d'una 'matriu' o 'taula'.

Per exemple:

```python
list_of_lists = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print(list_of_lists)
```

Per tal d'afegir files, o elements a la llista de llistes, podem utilitzar la funció append(). Per exemple:

```python
list_of_lists.append([3,2,1])
print(list_of_lists)
```

A vegades, podem no saber quins elements afegir a la llista. Per aquest motiu, podem 'inicialitzar' (crear una llista sense valors) una llista, i afegir elements de la següent manera:

```python
# Initializing an empty list 
list_of_lists = [] 
  
# Appending lists to create a list of lists 
list_of_lists.append([1, 2, 3]) 
list_of_lists.append([4, 5, 6]) 
list_of_lists.append([7, 8, 9]) 
  
# Displaying the resulting list of lists 
print(list_of_lists)
```

Per tal d'accedir a una fila en concret, podem fer-ho com ho havíem fet fins ara amb les llistes:

```python
fila0 = list_of_lists[0]
print(fila0)
```

Si volem accedir a un cert valor d'una fila, és a dir, especificar fila i columna (si ho veiem com una taula o matriu), podem fer-ho de la següent manera:

```python
fila = 0
columna = 2

print(list_of_lists[fila][columna])
```

Per accedir a una posició en concret, doncs, cal indicar mitjançant dos claudàtors primer, la fila i, després, la columna: **[fila][columna]**.

## Dictionaries

Un altre tipus de tipus de dades són els **diccionaris**. Són **mutables**, i la seva característica principal és que, a banda d'emmagatzemar els valors, també es poden classificar per categories, anomenades **unique keys** o **keys** directament. Les **keys** poden ser Int, Float o String.

Per exemple, el següent dictionary emmagatzema els elements i el seu nombre atòmic:

```python
elements = {"hydrogen": 1, "helium": 2, "carbon": 6}
print(elements)
```

Els dictionaries poden contenir qualsevol tipus de dades, i cada **key** pot contenir més d'un valor. Per exemple, el següent dictionary conté un recompte d'animals:

```python
animals = {'dogs': [20, 10, 15, 8, 32, 15], 'cats': [3,4,2,8,2,4], 'rabbits': [2, 3, 3], 'fish': [0.3, 0.5, 0.8, 0.3, 1]}
print(animals)
```

Per obtenir els valors d'una **key**, podem fer-ho de dues maneres: tal com ho féiem amb les llistes, a través de claudàtors, o bé amb la funció **get()**.

```python
print(elements["hydrogen"])
print(elements.get("hydrogen"))
```

Per modificar o afegir en un dictionary, podem fer-ho de la següent manera: 

```python
# Modificar un valor
elements["helium"] = 4
print(elements)
```

```python
# Afegir elements
elements["dilithium"] = 5
print(elements)
```

Per comprovar si una **key** forma part d'un dictionary, podem fer-ho utilitzant **in**. Per exemple:

```python
print("carbon" in elements)
print("oxygen" in elements)
```

Observeu la diferència entre utilitzar la cerca amb [] i get() per **keys** no existents:

```python
print(elements["oxygen"])
print(elements.get("oxygen"))
```
