# Strings

El text o String es poden expressar també a través de variables, indiferentment entre cometes senzilles (' ') o dobles (" ").

```python
my_string = 'this is a string!'
print(my_string)
my_string2 = "this is also a string!!!"
print(my_string2)
```

En cas de voler posar una coma simple entre mig del text, com ara per dir "L'avió del Nil", cal fer-ho afegint '\' prèviament a la cometa:

```python
this_string = 'L\'avió del Nil.'
print(this_string)
```

Perquè, sinó, ens apareix un error:

```python
this_string = 'L'avió del Nil.'
print(this_string)
```

Una altra operació freqüent és la de **concatenar** Strings, és a dir, ajuntar Strings. Per exemple:

```python
primera_paraula = "Hola"
segona_paraula = "classe"
print(primera_paraula+segona_paraula)
print(primera_paraula+" "+segona_paraula)
```

També podem escriure tantes vegades com vulguem una paraula tot multiplicant-la per un número:

```python
print("Hola" * 5)
print(segona_paraula * 10)
```

Amb Python existeixen molts tipus de **funcions** que ens faciliten la programació. Per exemple, en cas de voler veure quantes lletres té un text, podem fer-ho amb **len()** (*longitud* d'una string):

```python
print(len(primera_paraula))
print(len(segona_paraula))
```

Finalment, si volem veure quina és la lletra que hi ha en la posició n d'una paraula, podem fer-ho de la següent manera:

```python
n = 0
print(primera_paraula[n])
```

**IMPORTANT**: en Python, els comptes van de 0 a longitud - 1. Per exemple, si volem saber quina és la primera lletra d'una paraula, aquesta serà la posició 0. Si una paraula té N lletres, l'última lletra serà la N-1. Cal recordar això perquè serà útil més endavant.

```python
n = len(primera_paraula)
print(primera_paraula[n])
```
