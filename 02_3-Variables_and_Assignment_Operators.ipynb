# Variables and Assingment Operators

Imprimir-ho tot directament per pantalla pot ser incòmode. A més, no sempre necessitarem imprimir les coses per pantalla, i enlloc d'això voldrem guardar els resultats en algun lloc. Per aquest motiu, existeixen les **variables**. Així doncs, en Python (i qualsevol tipus de llenguatge) podrem crear variables i assignar-los-hi valors. Alhora, aquestes variables podran ser utilitzades (o no) més tard. 

Per exemple, crearem una variables que s'anomena 'var', i li assignarem el valor de 10. Per tal d'assignar valors, s'utilitza el símbol '='. Finalment, imprimirem el valor de la variable 'var' per pantalla:

```python
var = 10
print(var)
```

Com veieu, 'var' ha guardat el valor i després hem pogut imprimir-lo. Les variables poden ser anomenades com vulguem, però normalment:
- S'escriuen en mínuscules
- No es poden fer servir espais. Enlloc, cal fer servir '_'. Per exemple: *poblacio_riudoms*
- Es poden fer servir números per anomenar les variables. Per exemple: var1 o var_1.
- Algunes paraules no es poden fer servir com a nom de variables. Són les anomenades "keywords" o "Python's reserved words". Les podeu trobar a https://docs.python.org/3/reference/lexical_analysis.html#keywords.

Per exemple, les següents variables estan mal anomenades: my height = 58, MYLONG = 40 i MyLat = 105. En la primera, fins i tot, ens donaria un error en el codi:

```python
my height = 58
```

En el següent exemple, s'assignen valors a les variables x, y i z de dues maneres diferents. Comproveu que el resultat és el mateix (imprimiu els valors de cadascuna per pantalla)

```python
x = 3
y = 4
z = 5
```

```python
x, y, z = 3, 4, 5
```

Per imprimir en una mateixa frase text i el valor d'una variable, hem de fer dues coses:
1. Convertir la variable a tipus text (String). Per fer-ho, s'ha de fer: str(x)
2. Utilitzar el símbol '+' per afegir la variable al text.

Per exemple:

```python
print("x="+str(x)+"\ny="+str(y)+"\nz="+str(z))
```

Alternativament, també podem escriure una ',' i no serà necessari convertir a String. Tot i així, fixeu-vos que, en aquest cas, apareix un espai ' ' entre l'= i el resultat, que a vegades podem no voler.

```python
print("x=",x,"\ny=",y,"\nz=",z)
```

Amb les variables també podem operar. Per exemple:

```python
z = x + y + z

print(z)
```
