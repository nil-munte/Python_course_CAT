# DATA TYPES

Els **data types** o **tipus de dades** en Python són, tal com diu el nom, els tipus de dades que ens podem trobar. Aquests són universals per tots els llenguatges, és a dir, que us els trobareu en d'altres llenguatges de programació. Els que veurem en aquest curs són:
- **Enters (Integers o Int)**: són els números enters, com ara -2, -1, 0, 1, 2, 3, ...
- **Decimals (Float)**: són els números decimals, com ara -2.2, -1.0, 0, 1.0, 2.3, 3.456, ...
- **Booleans**: els booleans són aquell tipus de dades que només tenen dos valors, o bé *False* (o '0'), o bé *True* (o '1').
- **Strings**: són el que coneixem col·loquialment com text. Per exemple, "Hola" o "Avui fa bon dia." S'expressen entre cometes (" ").

# ARITHMETIC OPERATORS

Els **operadors bàsics** en Python són:
<br>(+) Suma
<br>(-) Resta
<br>(*) Multiplicació
<br>(/) Divisió
<br>(%) Mod (residu d'una divisió)
<br>(**) Potència
<br>(//) Divisió entera, és a dir, que trunca el resultat

### Exemples

```python
print(3 + 5)
```

```python
print(1 + 2 + 3 * 3)
```

```python
print(3 ** 2)
```

```python
print(9 % 2)
```
